- Using  dictionary comprehension to combine the two lists and calculate the sample means drawn from the dictionary with the inputs being the sample size and the dataset i.e., our dictionary
Plotting graphs (mean highlighted in each case):
 - Data plotted using the names asceding
 - Data plotted using the heights descending
 - Data plotted highlighting the maximum, minimum and median heights

```python
#Creating the lists
names = [
    "Alice", "Bob", "Charlie", "Diana", "Ethan",
    "Fiona", "George", "Hannah", "Isaac", "Julia",
    "Kevin", "Laura", "Michael", "Nina", "Oscar"
]
 
heights = [
    165, 178, 172, 160, 185,
    170, 182, 158, 174, 169,
    180, 162, 176, 168, 181
]
```

```python
#Importing the libraries
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import itertools
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
#converting the lists into a dictionary
names_dictionary = dict(zip(names, heights)) # combine the lists to form a dictionary
names_dictionary
```

```python
#Function for calculating  the mean
def calculate_mean(n):
    ndict_mean=sum(n.values())/len(n)
    return ndict_mean
```

```python
#Calculating the mean for the names dictionary
names_dictionary_mean=calculate_mean(names_dictionary)
names_dictionary_mean
```

```python
#Ploting the graph highlighting the mean
#Plotting the graph
plt.figure(figsize=(12,6)) # Increasing the size of the graph
plt.bar(names, heights, color="blue", edgecolor="black") # The bar colour

#Plotting the mean line
plt.axhline(names_dictionary_mean, color="green", linestyle="--", linewidth=3, label=f"Mean = {names_dictionary_mean}")

#Naming and bolding the axes and title
plt.xlabel("Names", fontweight="bold") # Labelling and bolding the x-axis
plt.ylabel("Heights (cm)",fontweight="bold") # Labelling and bolding the y-axis
plt.title("People's Heights Highlighting the Mean",fontweight="bold") # Labelling the graph and bolding the title

#Rotating the x-axis values for readability and displaying the graph
plt.xticks(rotation=45, ha="right")  
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Sort dictionary items by height (value) in descending order
sorted_heights = dict(sorted(names_dictionary.items(), key=lambda x: x[1], reverse=True))

print(sorted_heights)
```
